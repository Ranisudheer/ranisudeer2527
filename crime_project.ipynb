```python
#Q1 Most case reported city
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('E:\\Data Science\\Assignment\\crime_dataset_india.csv')

# and sum the most reported city
# Assume the payment method column is named 'payment_method'
most_reported_city = df['City'].value_counts().nlargest(5)

print("The most crime city is:", most_reported_city)

# Plot a bar chart to visualize the results
plt.figure(figsize=(12, 14))
plt.bar(most_reported_city.index, most_reported_city.values)
plt.title('No. of Reported city')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.show()

#most crime domain
most_reported_count = df['City'].value_counts().max(), most_reported_city.index[0]
print("The most crime city is:", most_reported_count)
```

```python
#Q2 Least reported crime city

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('E:\\Data Science\\Assignment\\crime_dataset_india.csv')

# and sum the Least reported city
# Assume the payment method column is named 'City'
least_reported_city = df['City'].value_counts().nsmallest(5)
# Display the resulting DataFrame

# Plot a bar chart to visualize the results
plt.figure(figsize=(12, 14))
plt.bar(least_reported_city.index, least_reported_city.values)
plt.title('No. of Reported city')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.show()

#least crime domain
least_reported_count = df['City'].value_counts().min(), least_reported_city.index[-1]
print("The least crime city is:", least_reported_count)
```

```python
#Q5 most and least case reported month

import pandas as pd
import datetime as date
#import matplotlib.pyplot as plt
df = pd.read_csv('E:\Data Science\Assignment\crime_dataset_india.csv')
dates = ['Date']  # list of dates in the format YYYY-MM-DD
def extract_month(date):
    """Extract month from datetime object"""
    return date.month
df['Date Reported'] = pd.to_datetime(df['Date Reported'])
#Extract months
months = df['Date Reported'].apply(extract_month)
# Count occurrences of each month
month_counts = months.value_counts()
#           """Adding month info to code"""
month_names = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}
#name  most_least_branch = reported.index[-1]
most_reported_month = month_counts.idxmax()
least_reported_month = month_counts.idxmin()

print(f"Most case reported month: {month_names[most_reported_month]}")
print(f"Least case reported month: {month_names[least_reported_month]}")

```

```python
#Q8 which time did the crime occured
#Time of Occurrence

import pandas as pd
import matplotlib.pyplot as plt
from datetime import date
df = pd.read_csv('E:\\Data Science\\Assignment\\crime_dataset_india.csv')

# Convert 'Time_of_Occurrence' to datetime
df['Time_of_Occurrence'] = pd.to_datetime(df['Time of Occurrence'])
#print(df)
# Extracting the time
df['Time_Only'] = df['Time_of_Occurrence'].dt.time
#print(df[['Time_Only']])

# Define a function to categorize the time of day
def categorize_time_of_day(Time_of_Occurrence):
    if Time_of_Occurrence.hour < 6:
        return 'Night'
    elif Time_of_Occurrence.hour < 12:
        return 'Morning'
    elif Time_of_Occurrence.hour < 18:
        return 'Afternoon'
    else:
        return 'Evening'
# Apply the function to create a new column
df['time_of_day'] = df['Time_of_Occurrence'].apply(categorize_time_of_day)

crime_counts = df['time_of_day'].value_counts()
print(crime_counts)


#most crime active time
most_crime_time = df['Time_of_Occurrence'].value_counts().max(), crime_counts.index[0]
print("The most reported crime time is:", most_crime_time)

#least crime active time
least_crime_time = df['Time_of_Occurrence'].value_counts().min(), crime_counts.index[-1]
print("The least reported crime tme is:", least_crime_time)



```

```python
#Q9 what age category is the victim 
#age category of the customers

import pandas as pd
import matplotlib.pyplot as plt
from datetime import date
df = pd.read_csv('E:\\Data Science\\Assignment\\crime_dataset_india.csv')

#4 - age category of the customers
# Assume the age column is named 'Age'
bins = [0, 12, 18, 35, 50, 100]  # Define age bins
labels = ['Child', 'Teen', 'Adult', 'Middle-aged', 'Senior']  # Define labels for each bin


# Create age categories using pd.cut
age_categories = pd.cut(df['Victim Age'], bins=bins, labels=labels)

# Get the count of each age category
age_category_counts = age_categories.value_counts()

# Print the count of each age category
print(age_category_counts)

# Plot a bar chart to visualize the results
plt.figure(figsize=(8, 6))
plt.bar(age_category_counts.index, age_category_counts.values)
plt.title('Age Categories of Victims')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.show()


```
